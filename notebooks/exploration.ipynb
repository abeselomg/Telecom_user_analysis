Mean vector and Covariance Matrix

```python
cleaned_numpy = aggregate_info.to_numpy()

mean_list = []

for i in range(aggregate_info.shape[0]):
    np.mean(cleaned_numpy[i, :])
    mean_list.append(np.mean(cleaned_numpy[i, :]))

mean_vector = np.array(mean_list)

print("Mean Vector:\n", mean_vector)

cov_mat = np.cov([cleaned_numpy[0, :], cleaned_numpy[1, :], cleaned_numpy[2, :]])
print("Covariance Matrix:\n", cov_mat)
```

```python
combined = aggregate_info[
    [
        "Total UL/DL",
        "Google UL/DL",
        "Youtube UL/DL",
        "Email UL/DL",
        "Gaming UL/DL",
        "Netflix UL/DL",
        "Social Media UL/DL",
        "Other UL/DL",
    ]
]
correlation_map = combined.corr()
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(correlation_map, dtype=np.bool))
heatmap = sns.heatmap(
    correlation_map, mask=mask, vmin=-1, vmax=1, annot=True, cmap="BrBG"
)
heatmap.set_title("Triangle Correlation Heatmap", fontdict={"fontsize": 18}, pad=16)
```
